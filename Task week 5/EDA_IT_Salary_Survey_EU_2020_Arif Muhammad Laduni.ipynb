prompt ChatGPT: Berikan code untuk mengubah fitur-fitur berikut "Total years of experience", "Years of experience in Germany","Yearly bonus + stocks in EUR", "Annual bonus+stocks one year ago. Only answer if staying in same country", "Number of vacation days". bertipe data numerik !

prompt ChatGPT: beri code untuk visualisasi data untuk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt

# Filter data berdasarkan rentang usia 20-69 tahun
filtered_data = df[(df['Age'] >= 20) & (df['Age'] <= 69)]

# Membuat histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_data['Age'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age (20-69)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Prompt ChatGPT : buatlah visualisasi yang berbeda setiap variabel untuk memahami distribusi dan hubungan antar variabel

```python
import seaborn as sns

# Histogram untuk variabel numerik
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Box plot untuk variabel numerik
sns.boxplot(x='Age', data=df)
plt.title('Boxplot of Age')
plt.show()

# Scatter plot untuk hubungan antara dua variabel numerik
sns.scatterplot(x='Age', y='Total experience', data=df)
plt.title('Scatter Plot of Age vs Total Experience')
plt.xlabel('Age')
plt.ylabel('Total Experience')
plt.show()


# Scatter plot untuk variabel kategorikal
sns.stripplot(x='Programming language', y='Age', data=df, jitter=True)
plt.title('Scatter Plot of Age by Programming Language')
plt.xlabel('Programming Language')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap untuk korelasi antar variabel numerik
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
