```python
# Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars_df = '/content/drive/MyDrive/Dataset/used_cars_data.csv'

# Membaca file CSV dan membuat DataFrame
cars_df = pd.read_csv(cars_df)

# Menampilkan beberapa baris pertama dari DataFrame
cars_df.head()
```

# Box Plot
prompt ChatGPt:berikan code box plot untuk fitur numerik

```python
# Memilih hanya kolom numerik
numeric_data = ['Year', 'Kilometers_Driven', 'Seats', 'Price']

# Membuat box plot untuk setiap fitur numerik
for feature in numeric_data:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=cars_df[feature], color='red')
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

# Scatter Plot
berikan code scatter plot price dan location

```python
# Membuat scatter plot antara price dan location
plt.figure(figsize=(10, 6))
plt.scatter(cars_df['Location'], cars_df['Price'], alpha=0.5)
plt.title('Scatter Plot Price vs Location')
plt.xlabel('Location')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Heat Map
berikan code heat map untuk fitur numerik



```python
# Memilih hanya fitur numerik
numeric_features = cars_df.select_dtypes(include=['int64', 'float64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Antar Fitur Numerik')
plt.tight_layout()
plt.show()
```

# Violin plot
berikan violin plot untuk setiap fitur numerik

```python
# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_features.columns):
    plt.subplot(2, 4, i + 1)
    sns.violinplot(x=cars_df[col], inner="quartile")
    plt.title(f'Violin Plot untuk {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

# Bar plot

berikan code bar plot untuk kilometer_driven berdasarkan tahun dengan warna yang berbeda


```python

# Membuat bar plot untuk kilometer_driven berdasarkan tahun
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Kilometers_Driven', data=cars_df, estimator=sum, palette='viridis')
plt.title('Total Kilometer Driven Berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Kilometer Driven')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
